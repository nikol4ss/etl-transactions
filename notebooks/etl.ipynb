```python
import pandas as pd
```

```python
df = pd.read_csv("../data/transactions.csv")
df.dtypes
```

```python
df['descricao'] = df['descricao'].fillna('No description')
invalid_values = df[df.isnull().any(axis=1) | (df['valor'] < '0') | (df['valor'] == 'erro_valor')]
df = df.drop(invalid_values.index)
```

```python
invalid_date = df[pd.to_datetime(df['data'], errors='coerce').isna()].copy()
df = df.drop(invalid_date.index)
```

```python
invalid_values = pd.concat([invalid_values, invalid_date], ignore_index=True)
```

```python
df['valor'] = df['valor'].astype(float)
df['tipo'] = df['tipo'].astype('category')
df['status'] = df['status'].astype('category')
df['data'] = pd.to_datetime(df['data'])
```
